```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import RobustScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor
import warnings
warnings.filterwarnings('ignore')


df = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')


df.head()
```

<h2 dir=rtl align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
ایجاد متغیر هدف
</font>
</h2>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در این قسمت باید متغیر هدف مسئله را ایجاد کنیم. در این مسئله هدف ما تخمین ارزش خانه‌های مورد نظر است. برای این منظور، با استفاده از فرمول زیر، ستونی با نام <code>Total</code> به مجموعه داده خود اضافه کنید تا مدل یادگیری ماشین خود را منطبق بر آن آموزش دهید.

$$ Total = Credit + (Rent * 33.33334)$$

</font>
</p>

```python

df['Total'] = df['Credit'] + (df['Rent'] * 33.33334)

```

delete outlier

```python
lower_bound = df['Total'].quantile(0.01)
upper_bound = df['Total'].quantile(0.99)
df_clean = df[(df['Total'] >= lower_bound) & (df['Total'] <= upper_bound)]

# Log transform target 
df_clean['Total_log'] = np.log1p(df_clean['Total'])

print(f"\nLog transformed target created")
```

cleaning data

```python
def engineer_features(data, is_train=True, encoders=None):
   
    df_feat = data.copy()
    
    
    numeric_cols = ['Area', 'Year_built', 'Rooms', 'Units_per_floor']
    for col in numeric_cols:
        if col in df_feat.columns:
            df_feat[col] = pd.to_numeric(df_feat[col], errors='coerce')
    
   
    if 'Floor' in df_feat.columns:
        df_feat['Floor_clean'] = df_feat['Floor'].astype(str)
        
        df_feat['Floor_Num'] = df_feat['Floor_clean'].str.extract(r'(\d+)')[0]
        df_feat['Floor_Num'] = pd.to_numeric(df_feat['Floor_Num'], errors='coerce')
      
        df_feat['Floor_Num'] = df_feat['Floor_clean'].apply(
            lambda x: 0 if 'همکف' in str(x) else x
        )
        df_feat['Floor_Num'] = pd.to_numeric(df_feat['Floor_Num'], errors='coerce')
        
        df_feat['Floor_Total'] = df_feat['Floor_clean'].str.extract(r'از\s*(\d+)')[0]
        df_feat['Floor_Total'] = pd.to_numeric(df_feat['Floor_Total'], errors='coerce')
    
    binary_cols = ['Elevator', 'Parking', 'StorageRoom', 'Balcony']
    for col in binary_cols:
        if col in df_feat.columns:
            df_feat[col] = df_feat[col].astype(str)
            df_feat[col] = df_feat[col].map({'دارد': 1, 'ندارد': 0})
            df_feat[col] = df_feat[col].fillna(0)
    

    if 'Year_built' in df_feat.columns:
        current_year = 2026
        df_feat['Building_Age'] = current_year - df_feat['Year_built']
        df_feat['Building_Age'] = df_feat['Building_Age'].clip(0, 100)
        df_feat['Is_New'] = (df_feat['Building_Age'] <= 5).astype(int)
        df_feat['Is_Old'] = (df_feat['Building_Age'] >= 30).astype(int)
    
   
    if 'Area' in df_feat.columns and 'Rooms' in df_feat.columns:
        df_feat['Area_per_Room'] = df_feat['Area'] / df_feat['Rooms']
        df_feat['Area_per_Room'] = df_feat['Area_per_Room'].replace([np.inf, -np.inf], np.nan)
    
   
    if 'Floor_Num' in df_feat.columns and 'Floor_Total' in df_feat.columns:
        df_feat['Floor_Ratio'] = df_feat['Floor_Num'] / df_feat['Floor_Total']
        df_feat['Floor_Ratio'] = df_feat['Floor_Ratio'].replace([np.inf, -np.inf], np.nan)
        df_feat['Is_Ground'] = (df_feat['Floor_Num'] == 0).astype(int)
        df_feat['Is_Top'] = ((df_feat['Floor_Num'] == df_feat['Floor_Total']) & 
                             (df_feat['Floor_Total'].notna())).astype(int)
    

    if 'Area' in df_feat.columns:
        df_feat['Area_Squared'] = df_feat['Area'] ** 2
        if 'Elevator' in df_feat.columns:
            df_feat['Area_Elevator'] = df_feat['Area'] * df_feat['Elevator']
        if 'Parking' in df_feat.columns:
            df_feat['Area_Parking'] = df_feat['Area'] * df_feat['Parking']
    
    
    cat_cols = ['Rent_type', 'Building_direction', 'Unit_status', 
                'Floor_material', 'Bathroom', 'Heating', 'Cooling', 
                'Hot_water_provider', 'District', 'city']
    
    for col in cat_cols:
        if col in df_feat.columns:
            if is_train:
          
                freq_map = df_feat[col].value_counts(normalize=True).to_dict()
                df_feat[col + '_enc'] = df_feat[col].map(freq_map)
                if encoders is not None:
                    encoders[col] = freq_map
            else:
              
                if encoders and col in encoders:
                    df_feat[col + '_enc'] = df_feat[col].map(encoders[col])
                    df_feat[col + '_enc'] = df_feat[col + '_enc'].fillna(0)
    
    return df_feat, encoders
```

```python

encoders = {}
df_engineered, encoders = engineer_features(df_clean, is_train=True, encoders=encoders)


test_engineered, _ = engineer_features(test, is_train=False, encoders=encoders)
```

```python

feature_cols = [
    'Area', 'Rooms', 'Units_per_floor',
    'Building_Age', 'Is_New', 'Is_Old',
    'Elevator', 'Parking', 'StorageRoom', 'Balcony',
    'Area_per_Room', 'Area_Squared', 'Area_Elevator', 'Area_Parking',
    'Floor_Num', 'Floor_Total', 'Floor_Ratio', 'Is_Ground', 'Is_Top'
]

encoded_cols = [col for col in df_engineered.columns if col.endswith('_enc')]
feature_cols.extend(encoded_cols)


feature_cols = [col for col in feature_cols if col in df_engineered.columns]


```

```python

X = df_engineered[feature_cols].copy()
y = df_engineered['Total_log'].copy()

```

```python

imputer = SimpleImputer(strategy='median')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=feature_cols)
test_imputed = pd.DataFrame(imputer.transform(test_engineered[feature_cols]), columns=feature_cols)
```

```python

X_train, X_val, y_train, y_val = train_test_split(
    X_imputed, y, test_size=0.2, random_state=42
)

```

```python

scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
test_scaled = scaler.transform(test_imputed)
```

```python


rf = RandomForestRegressor(
    n_estimators=300,      # Number of trees
    max_depth=20,          # Maximum depth of each tree
    min_samples_split=10,  # Minimum samples to split a node
    min_samples_leaf=5,    # Minimum samples in a leaf node
    random_state=42,       # For reproducibility
    n_jobs=-1              # Use all CPU cores
)

rf.fit(X_train_scaled, y_train)
print("training completed")
```

r2 score 

```python

y_val_pred_log = rf.predict(X_val_scaled)

#back from log to original
y_val_pred = np.expm1(y_val_pred_log)
y_val_actual = np.expm1(y_val)


r2 = r2_score(y_val_actual, y_val_pred)
rmse = np.sqrt(mean_squared_error(y_val_actual, y_val_pred))


print(f"R² Score: {r2:.4f}    ")

```

This project develops a machine learning model to predict housing prices in Iran using a dataset of 382,467 training records. Key steps include creating the target variable Total = Credit + (Rent × 33.33334), applying log transformation to handle skewed price distribution, and engineering 20+ features such as Building_Age, Area_per_Room, and Floor_Ratio. Target encoding with smoothing was used for categorical variables like city and district. A Random Forest Regressor achieved an R² score of 0.81 on validation data, with area and location identified as the most important price predictors.
